```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = ["FW", "LC", "LIB",  "SC", "SPV", "Wind"]  # technology file names

df_set = {}

# read the data
for n in names:
    try:
        df_n = pd.read_csv(f"entropy_data/cond_entropy_{n}.csv")
        df_n['Publn_Date'] = pd.to_datetime(df_n['Publn_Date'])
        df_set[n] = df_n
    except:
        pass
```

```python
# plot full the data
for n in df_set.keys():
    df_set[n].plot(x='Publn_Date', y='Entropy', kind='line', marker='o', title=f'Total Entropy for {n} patents')
    plt.show()
```

```python
# plot for first x years the data
x = 2
for n in df_set.keys():
    df_set[n].plot(x='Publn_Date', y='Entropy', kind='line', marker='o', title=f'Total Entropy for {n} patents, first {x} years')
    start = df_set[n]['Publn_Date'].values[0]
    end = start + pd.DateOffset(years=x)
    plt.xlim(start, end)
    plt.show()
```
